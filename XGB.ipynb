```python
df = pd.merge(exercise,calorie,how='left',on = 'User_ID')
```

```python
df.info()
```

```python
df.head()
```

```python
X = df.drop(['User_ID','Calories'],axis=1)
```

```python
X.values
```

```python
y = df['Calories']
```

```python
y.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
```

```python
predict
```

```python
y_test
```

```python
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test,predict)
```

```python
print("Mean Absolute Error = ",mae)
```

```python
y_min = 1
y_max = 314
mae = 1.5211099035541216

# Calculate the percentage of the target variable range
percentage_error = (mae / (y_max - y_min)) * 100
print(f"MAE as a percentage of the target variable range: {percentage_error:.2f}%")
```
